```python
import sqlite3
import random
import time
import datetime
import matplotlib.pyplot as plt
%matplotlib notebook

# Criando uma conexão
conn = sqlite3.connect("dsa.db")

# Criando cursor
c = conn.cursor()

# Define uma função para criar uma tabela
def create_table():
    c.execute('CREATE TABLE IF NOT EXISTS produtos (id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL, date TEXT, '\
              'prod_name TEXT, valor REAL) ')

# Define uma função para criar uma linha    
def data_insert():
    c.execute("INSERT INTO produtos VALUES(10, '2018-05-02 14:32:11', 'Teclado', 130)")
    conn.commit()
    c.close()
    conn.close()

# Função para adicionar dados com variaveis    
def data_insert_var():
    new_date = datetime.datetime.now()
    new_prod_name = 'Monitor'
    new_valor = random.randrange(50,100)
    c.execute("INSERT INTO produtos(date, prod_name, valor) VALUES (?, ?, ?)",(new_date, new_prod_name, new_valor))
    conn.commit()
    
#Leitura dos dados
def leitura_todos_dados():
    c.execute("SELECT * FROM PRODUTOS")
    for linha in c.fetchall():
        print(linha)

#Leitura de registros específicos
def leitura_registros():
    c.execute("SELECT * FROM PRODUTOS WHERE valor > 60.0")
    for linha in c.fetchall():
        print(linha)

#Leitura de colunas específicas
def leitura_colunas():
    c.execute("SELECT * FROM PRODUTOS")
    for linha in c.fetchall():
        print(linha[3])

#Update
def atualiza_dados():
    c.execute("UPDATE produtos SET valor = 70.0 where valor = 82.0")
    conn.commit()
    
#Delete
def remove_dados():
    c.execute("DELETE FROM produtos WHERE valor = 60.0")
    conn.commit()
    
#Gerar Grafico com os dados do banco de dados
def dados_grafico():
    c.execute("SELECT id, valor FROM produtos")
    ids = []
    valores = []
    dados = c.fetchall()
    for linha in dados:
        ids.append(linha[0])
        valores.append(linha[1])
    
    plt.bar(ids, valores)
    plt.show()
```
